```python
def fakedata():
    import pandas as pd
    from faker import Faker
    import random
    fake = Faker()

    def persons():
        def per():        
            letters = 'bcdfghjklmnpqrstvwxyz'
            vowels = 'aeiou'
            df = pd.DataFrame(columns = ['first_name', 'last_name', 'phone_number', 'customer_number','credit'])
            df.name = 'persons'
            def lname():
                name = fake.last_name()
                for x in name[-1]:
                    if x =='y': pass
                    if x in letters:
                        name = name+random.choice(vowels)
                    else: name = name+random.choice(letters)
                if name[0] in 'BDPC'and (random.randint(0,3)==2):
                    name = 'Mc'+name
                if name[0] in 'A' and (random.randint(0,3)==1):
                    name = 'De'+name
                return name

            def ccfull():
                cclst = []
                for x in range(1000):
                    cclst.append(fake.credit_card_full())
                cc = pd.DataFrame(cclst)
                cc = cc[cc[0].apply(lambda x: 'Diners Club' not in x)]
                cc = cc[cc[0].apply(lambda x: ('VISA' in x) | ('JCB' in x))]
                cc_out = cc[0].iloc[random.randint(0,99)]
                return cc_out

            for x in range(100):
                df = df.append({'first_name': fake.first_name(), 'last_name':(lname()),
                                'phone_number': fake.phone_number(),'customer_number': fake.random_number(5),
                               'credit':ccfull()}, 
                               ignore_index = True)
            df = df.drop(index=0)
            df['first_name'].iloc[98] = 'Wayne'
            df['last_name'].iloc[98] = 'Gretzky'
            df['phone_number'].iloc[98] = '805-444-4564'
            df['customer_number'].iloc[98] = 99 #1979-1999, all hail the great one!! 1+9+7+9=26, 1+9+9+9=28, born 1/26 and 3/28 scores 92nd goal of the season. Coincidence? Maybe, lets run a ttest!

            df['credit'] = df['credit'].replace("[A-Z][a-z]+\s[A-z]?[a-z]+",df['first_name']+' '+df['last_name'],regex=True)

            return df
        out = per()
     
        while (  (len(out['customer_number'].unique()) < 99) or (len(out['last_name'].unique()) <99 )  )==True:
            out = per() 
        return out

    dfper = persons()

    def purchases():

        import pandas as pd
        from faker import Faker
        import random
        fake = Faker()

        df = pd.DataFrame(columns = ['isbn10', 'credit_card_full', 'price', 'customer_number'])
        df.name = 'purchases'

        def isbn10():
            Faker.seed(0)
            isblst = []
            for x in range(1000): isblst.append(fake.isbn10())
            isbn = pd.DataFrame(isblst)
            isbn_out = isbn[0].iloc[random.randint(0,999)]
            return isbn_out
        
        cc_dict = pd.Series(dfper['credit'].values,index=dfper['customer_number']).to_dict() 

        
        for x in range(99):
            df = df.append({'isbn10': isbn10(), 'credit_card_full':None,
                            'price': round(random.uniform(100, 1000),2),'customer_number': dfper['customer_number'].iloc[random.randint(0,98)]}, 
                           ignore_index = True)

        df['credit_card_full'] = df['customer_number'].apply(lambda x: cc_dict[x] if (x in cc_dict.keys()) else x)
            
        return df
    
    dfpur = purchases()
    
    dfper = dfper.drop('credit', axis=1)

    return dfper, dfpur
```

```python
dfper, dfpur = fakedata()
```
